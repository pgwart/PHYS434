```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

### Problem 0.

```python
t = np.linspace(0,np.pi*100,int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
```

### Problem 1.
#### 1A)

```python
plt.scatter(t, bkgd)
plt.show()
```

#### 1B)

```python
t0 = 500
n = 2000

plt.scatter(t[t0:t0+n], bkgd[t0:t0+n])
plt.show()
```

#### 1C)

```python
d = 100

plt.scatter(t[::d], bkgd[::d])
plt.show()
```

### Problem 2.
#### 2A)

```python
binsize = 100
plt.hist2d(t, bkgd, bins=binsize)
plt.colorbar()
plt.show()
```

#### 2B)
Instead of plotting each point individually, this plot takes all of the points in an area, which is defined by the bin size, and turns them into one point, where the color represents how many points are in that bin.

### Problem 3.
#### 3A)

```python
period = 2*np.pi
tfold = t%period
```

```python
d = 150
plt.scatter(tfold[::d], bkgd[::d])
plt.show()
```

#### 3B)

```python
plt.hist2d(tfold, bkgd, bins=100)
plt.colorbar()
plt.show()
```

#### 3C)

```python
bins=100
h, xedges, yedges, _ = plt.hist2d(tfold, bkgd, bins=bins)
ave = []
for i in h:
    total = 0
    for n in range(len(i)):
            total+=n*i[n]
    ave.append(total/len(i)/50)
plt.plot(xedges[1:], ave, c='r', label='Average')
colorbar = plt.colorbar()
plt.legend()
plt.show()
```
