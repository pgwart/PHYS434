#### 2A)

Using the formula for convolving two distributions, we get the probability density of the sum of two exponentional functions as $f_z(z) = \int_0^z \lambda_1 e^{-\lambda_1 x} \lambda_2 e^{\lambda_2 (z-x)} dx =\lambda^2 ze^{-\lambda z}$ when $\lambda_1 = \lambda_2$

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10,1000)

# Distributions where lambda = 1
plt.plot(x, np.exp(-x), label='One distribution')
plt.plot(x, x*np.exp(-x), label='Sum of two distributions')
plt.legend()
```

#### 2B)

Looking for a distribution that is the ratio of two normal distributions, I found the ratio distribution, and the Cauchy distribution, which is the case of the ratio distribution that is normal distributions with a mean of 0. The Cauchy distribution is defined as $f(x,\gamma) = { 1 \over \pi \gamma } \left[ { \gamma^2 \over (x)^2 + \gamma^2  } \right]$ for a distribution with a mean of 0, where $\gamma$ is the scale factor.

```python
from scipy.stats import norm
x = np.linspace(-10,10,1000)    
def cauchy(x, gamma):
    return (1/np.pi)*(gamma**2 / (x**2 + gamma**2))

plt.plot(x, norm.pdf(x), label='Gaussian')
plt.plot(x, cauchy(x, 1), label='Cauchy')
plt.legend()
plt.show()
```
